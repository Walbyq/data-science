## Тепловая карта по матрице корреляции

Корреляция указывает на степень взаимосвязи между двумя переменными. Она помогает понять, как одна переменная изменяется по отношению к другой. Значение варьируется от -1 до 1, где:

*   1 указывает на полную положительную корреляцию;
*   -1 указывает на полную отрицательную корреляцию;
*   0 указывает на отсутствие корреляции.





```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
```

## Гистограмма

Показывает, насколько населен регион по сравнению с его размерами.

```python
sns.histplot(df['Density (per km²)'], bins=30, kde=True)
plt.title('Распределение плотности населения')
plt.show()
```

## Диаграмма рассеяния

Позволяет увидеть соотношение между населением стран в 1970 и 2022 годах.

```python
sns.scatterplot(df, x='1970 Population', y='2022 Population', hue='Continent')
plt.title('Сравнение населения стран в 1970 и 2022 годах')
plt.show()
```

## Сводные диаграммы

Показывает изменение численности населения для 10 стран с наибольшим процентом от мирового населения за период с 1970 по 2022 год.

```python
years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population',
         '2010 Population', '2015 Population', '2020 Population', '2022 Population']

fig, axes = plt.subplots(4, 2, figsize=(10, 20))
fig.suptitle(f'Сравнение населения 10 стран в разных годах', fontsize=16, y=1)

top_10 = df.nlargest(10, 'World Population Percentage')
for i, year in enumerate(years):
    ax = axes[i // 2, i % 2]
    sns.barplot(x='Country/Territory', y=year, data=top_10, ax=ax)
    ax.set_title(f'Население в {year[:4]} году')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Ящик с усами

Показывает распределение процентного соотношения населения стран по континентам.

```python
sns.boxplot(df, x='Continent', y='Growth Rate')
plt.xticks(rotation=45)
plt.title('Процентное соотношение населения')
plt.show()
```
